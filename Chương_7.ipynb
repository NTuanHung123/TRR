# ĐỒ THỊ VÔ HƯỚNG ĐƠN



a. Đề bài: Vẽ đồ thị vô hướng đơn với: Đỉnh: {1, 2, 3, 4} — Cạnh: {(1,2), (1,3), (2,4)}

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 4)])
nx.draw(G, with_labels=True, node_color='lightblue')
plt.title("Đồ thị vô hướng đơn")
plt.show()

```

b. Yêu cầu sinh viên sửa và lý giải

•	Thêm đỉnh 5 và cạnh 2–5 → Quan sát thay đổi liên thông

•	Xóa cạnh 1–2 → Có đỉnh nào cô lập không?

•	Thêm cạnh 3–4 → Có chu trình trực quan không?

•	Đếm bậc đỉnh bằng cách nhìn số cạnh nối


```python
import networkx as nx
import matplotlib.pyplot as plt

# 1. Vẽ đồ thị ban đầu
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 4)])

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
nx.draw(G, with_labels=True, node_color='lightblue', node_size=800, font_size=12)
plt.title("Đồ thị ban đầu")

# 2. Thêm đỉnh 5 và cạnh 2–5
G1 = G.copy()
G1.add_edge(2, 5)

plt.subplot(2, 2, 2)
nx.draw(G1, with_labels=True, node_color='lightgreen', node_size=800, font_size=12)
plt.title("Thêm đỉnh 5 và cạnh (2,5)")

# 3. Xóa cạnh (1–2)
G2 = G1.copy()
G2.remove_edge(1, 2)

plt.subplot(2, 2, 3)
nx.draw(G2, with_labels=True, node_color='lightyellow', node_size=800, font_size=12)
plt.title("Xóa cạnh (1,2)")

# 4. Thêm cạnh (3–4)
G3 = G2.copy()
G3.add_edge(3, 4)

plt.subplot(2, 2, 4)
nx.draw(G3, with_labels=True, node_color='lightpink', node_size=800, font_size=12)
plt.title("Thêm cạnh (3,4)")

plt.tight_layout()
plt.show()

# 5. Đếm bậc của từng đỉnh
print("Bậc của từng đỉnh:")
for node, degree in G3.degree():
    print(f"Đỉnh {node}: bậc {degree}")
```

#ĐỒ THỊ CÓ HƯỚNG

a. Đề bài: Cho đồ thị có hướng. Đỉnh: {A, B, C, D} — Cạnh: (A→B), (A→C), (B→D), (C→D)

```python
G = nx.DiGraph()
G.add_edges_from([("A", "B"), ("A", "C"), ("B", "D"), ("C", "D")])
nx.draw(G, with_labels=True, node_color='lightgreen', arrows=True)
plt.title("Đồ thị có hướng")
plt.show()
```

b. Yêu cầu sinh viên sửa và lý giải

•	Thêm cạnh D→A → Kiểm tra đồ thị có chu trình không.

•	In G.successors("A") và G.predecessors("D") để phân tích đầu ra.

•	Đổi hướng cạnh (A→C) thành (C→A) → Quan sát lại sơ đồ.

•	Kiểm tra tính DAG bằng nx.is_directed_acyclic_graph(G).


```python
import networkx as nx
import matplotlib.pyplot as plt

# 1. Vẽ đồ thị ban đầu
G = nx.DiGraph()
G.add_edges_from([("A", "B"), ("A", "C"), ("B", "D"), ("C", "D")])

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
nx.draw(G, with_labels=True, node_color='lightgreen', node_size=800, font_size=12, arrows=True)
plt.title("Đồ thị ban đầu")

# 2. Thêm cạnh D→A
G1 = G.copy()
G1.add_edge("D", "A")

plt.subplot(2, 2, 2)
nx.draw(G1, with_labels=True, node_color='lightblue', node_size=800, font_size=12, arrows=True)
plt.title("Thêm cạnh D→A (có chu trình)")

# 3. In successors và predecessors
print("Successors của A:", list(G1.successors("A")))
print("Predecessors của D:", list(G1.predecessors("D")))

# 4. Đổi hướng cạnh (A→C) thành (C→A)
G2 = G1.copy()
G2.remove_edge("A", "C")
G2.add_edge("C", "A")

plt.subplot(2, 2, 3)
nx.draw(G2, with_labels=True, node_color='lightyellow', node_size=800, font_size=12, arrows=True)
plt.title("Đổi hướng cạnh (A→C) thành (C→A)")

# 5. Kiểm tra DAG
is_dag = nx.is_directed_acyclic_graph(G2)
print("Đồ thị G2 là DAG?", is_dag)

plt.tight_layout()
plt.show()
```

#ĐA ĐỒ THỊ


a. Đề bài: Vẽ đa đồ thị có Đỉnh: {1, 2, 3} — Cạnh: (1,2), (1,2), (2,3)

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.MultiGraph()
G.add_edges_from([(1, 2), (1, 2), (2, 3)])
pos = nx.spring_layout(G)

plt.figure(figsize=(6, 4))
nx.draw_networkx_nodes(G, pos, node_color='orange')
nx.draw_networkx_labels(G, pos)

# Vẽ cạnh phân biệt đa cạnh
for i, (u, v, k) in enumerate(G.edges(keys=True)):
    nx.draw_networkx_edges(
        G, pos, edgelist=[(u, v)],
        connectionstyle=f'arc3, rad={(0.2 if k % 2 == 0 else -0.2)}',
        edge_color='black'
    )

plt.title("Đa đồ thị")
plt.axis('off')
plt.show()
```

b. Yêu cầu sinh viên sửa và lý giải

•	Thêm cạnh thứ 3 giữa 1 và 2 → Quan sát số cạnh thực tế.

•	Đổi MultiGraph sang Graph và thử thêm cạnh trùng → Lỗi hay ghi đè?

•	Thêm cạnh khuyên tại đỉnh 2 → So sánh G.degree(2) trước và sau.

•	Tạo lại bằng DiGraph → Có còn là đa đồ thị không?


```python
import networkx as nx
import matplotlib.pyplot as plt

# 1. Tạo đa đồ thị ban đầu
G = nx.MultiGraph()
G.add_edges_from([(1, 2), (1, 2), (2, 3)])
pos = nx.spring_layout(G, seed=42)

plt.figure(figsize=(12, 8))

# Vẽ hàm riêng cho MultiGraph
def draw_multigraph(G, pos, subplot, title):
    plt.subplot(2, 2, subplot)
    nx.draw_networkx_nodes(G, pos, node_color='orange', node_size=700)
    nx.draw_networkx_labels(G, pos)
    for i, (u, v, k) in enumerate(G.edges(keys=True)):
        nx.draw_networkx_edges(
            G, pos, edgelist=[(u, v)],
            connectionstyle=f'arc3, rad={(0.2 if k % 2 == 0 else -0.2)}',
            edge_color='black'
        )
    plt.title(title)
    plt.axis('off')

# Đồ thị ban đầu
draw_multigraph(G, pos, 1, "Đa đồ thị ban đầu")

# 2. Thêm cạnh thứ 3 giữa 1 và 2
G.add_edge(1, 2)
draw_multigraph(G, pos, 2, "Thêm cạnh thứ 3 giữa 1 và 2\n→ Có 3 cạnh (1,2)")

# 3. Chuyển sang Graph thường và thử thêm cạnh trùng
G_simple = nx.Graph()
G_simple.add_edges_from([(1, 2), (1, 2), (2, 3)])  # cạnh trùng bị ghi đè
plt.subplot(2, 2, 3)
nx.draw(G_simple, pos, with_labels=True, node_color='lightblue', node_size=700)
plt.title("Graph thường (cạnh trùng bị ghi đè)")

# 4. Thêm cạnh khuyên tại đỉnh 2
before = G.degree(2)
G.add_edge(2, 2)  # vòng khuyên
after = G.degree(2)
print("Bậc của đỉnh 2 trước khi thêm khuyên:", before)
print("Bậc của đỉnh 2 sau khi thêm khuyên:", after)

draw_multigraph(G, pos, 4, "Thêm cạnh khuyên tại đỉnh 2")

plt.tight_layout()
plt.show()

# 5. Tạo lại bằng DiGraph
DG = nx.DiGraph()
DG.add_edges_from([(1, 2), (1, 2), (2, 3)])  # cạnh trùng không tồn tại
print("DiGraph có phải đa đồ thị không?", isinstance(DG, nx.MultiDiGraph))
```
